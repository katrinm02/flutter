```
# Поиск всевозможных вариации таговвого номера
def get_tg_number(tg_n):
  l_tg = tg_n.split('/')

  ind = l_tg.index(max(l_tg, key=len))
  if ind == 0:
    main = l_tg[ind]
    list_ = [main]
    for i in l_tg[1:]:
      main = main[:-len(i)] + i
      list_.append(main)
  # else:
  elif ind == len(l_tg)-1:
    main = l_tg[ind]
    list_ = [main]
    for i in l_tg[:ind]:
      main = i + main[len(i):]
      list_.append(main)
  else:
    main = l_tg[ind]
    list_ = [main]
    for i in l_tg[ind:]:
      print(i)
      print(main[:-len(i)] + i)
    print(l_tg)
  return '/'.join(list_)
```

```
def from_rus_to_en(string):
  slovar = {
  'А':'A',
  'В':'B',
  'С':'C',
  'Е':'E',
  'Н':'H',
  'К':'K',
  'М':'M',
  'О':'O',
  'Р':'P',
  'Т':'T',
  'Х':'X'
  } 
  for key in slovar.keys():
      string = str(string).upper().replace(key, slovar[key])
  return string
```

```
def sc2(massive_1_, massive_4_):
  massive_1 = massive_1_
  massive_1['кол инв'] = massive_1['Номера инвойсов без даты из К'].map(lambda x: len(str(x).split('/')))
  massive_1 = massive_1[massive_1['кол инв'] == 1]

  massive_4 = massive_4_
  # column = 'Комментарий.1'
  column = '0'
  massive_4['инвойс'] = massive_4['НомерВходящегоДокумента'].map(lambda x:str(x).split('/')[0].split(' ')[0])
  massive_4 = massive_4[massive_4.loc[:, column:'17'].count(axis='columns') == 1]
  result = pd.merge(massive_1, massive_4, how='inner', left_on='Номера инвойсов без даты из К', right_on='инвойс')
  for ind, row in result[result['Артикул'].notna()][result['Таговый номер из E'].notna()].iterrows():
    if row['Артикул'] != row['Таговый номер из E'] or row['Артикул'] not in row['Таговый номер из E']:
      result.drop(ind, inplace=True)
  return result
```

```
def sc3(massive_1_, massive_4_):
  massive_1 = massive_1_[massive_1_['Таговый номер из E'].notna()]
  massive_1['Таговый номер'] = massive_1['Таговый номер из E'].map(lambda x: re.sub(r'[^a-zA-Z0-9_\/]', r'', str(x)) if pd.notna(x) else x)

  massive_4 = massive_4_[massive_4_['Артикул'].notna()]
  massive_4['Таговый номер'] = massive_4['Артикул'].map(lambda x: re.sub(r'[^a-zA-Z0-9_\/]', r'', str(x)).split('/'))
  # column = 'Комментарий.1'
  column = '0'
  if not massive_1.empty and not massive_4.empty:

    join_map = {}
    for ind,row in massive_1.iterrows():
      val = row['Таговый номер']
      if pd.notna(val):
        for i in val.split('/'):
          if i in join_map.keys():
            join_map[i].append(ind)
          else:
            join_map[i] = [ind]

    for key in join_map:
      if len(join_map[key]) == 1:
        pass
      else:
        dict_dt = {massive_1['№ ТД'][i]:i for i in join_map[key]}
        join_map[key] = dict_dt
    # print(join_map)

    def f(x):
      for i in massive_4['Таговый номер'][x]:
        if i in join_map.keys():
          val1 = join_map[i]
          if type(val1) == list:
            return val1[0]
          else:
            dt = massive_4[column][x]
            if dt in val1.keys():
              return join_map[i][dt]
            else:
              return 'не тот номер дт'

    massive_4['ind_row'] = massive_4.index.map(lambda x:f(x)).astype('object')
    massive_1.index = massive_1.index.astype('object')
    b_result = pd.merge(massive_1, massive_4, left_index=True, right_on='ind_row', how='inner')
  else:
    b_result = pd.DataFrame()
  # b_result['dt'] = b_result.index.map(lambda x : '+' if b_result.loc[x, '№ ТД'] in [str(i) for i in b_result.loc[x, column:'8'] if pd.notna(i)] else '-') #:12
  return b_result
```

```
def get_map(dataf, str_map):
  for ind, row in dataf.iterrows():
    val = row['номер строки_x']
    if val not in str_map:
      str_map.append(val)
  return str_map 

def get_map_mas4(dataf, str_map_mas4):
  for ind, row in dataf.iterrows():
    val = row['номер строки_y']
    if val not in str_map_mas4:
      str_map_mas4.append(val)
  return str_map_mas4

def get_empty(massive_1_, massive_4, scs):
  str_map = []
  str_map_mas4 = []

  for i in range(len(scs)):
    str_map = get_map(scs[i], str_map)
    str_map_mas4 = get_map_mas4(scs[i], str_map_mas4)

  massive_4_not_found = massive_4[~massive_4['номер строки_y'].isin(str_map_mas4)]
  massive_1_not_found = massive_1_[~massive_1_['номер строки_x'].isin(str_map)]

  return massive_1_not_found, massive_4_not_found

def to_view(table):
  s_m_3 = {}
  for ind, row in table.iterrows():
    val = table.loc[ind, 'номер строки_x']
    if val in s_m_3.keys():
      if row['номер строки_y'] not in s_m_3[val][0]:
        s_m_3[val][0].append(row['номер строки_y'])
    else:
      s_m_3[val] = [[row['номер строки_y']]]
  result_sc_ch = pd.DataFrame.from_dict(s_m_3, orient='index').reset_index()
  result_sc_ch.columns = ['0_x', '0_y']
  result_sc_ch['0_x'] = result_sc_ch['0_x'].map(lambda x: [x])
  return result_sc_ch
```

```
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("russian")

import pymorphy2

morph = pymorphy2.MorphAnalyzer()

def get_word(word):
  word = morph.parse(word)[0]
  return word.make_agree_with_number(1).word

def get_keywords(df, column_name, table):
  all_ = {}
  df_ = df

  for i in range(len(df_[column_name])):
    if pd.isna(df_[column_name][i]):
      continue
    elif table == 1:
      n_s = df_['номер строки_x'][i]
    if table == 2:
      n_s = df_['номер строки_y'][i]
    all_[n_s] = list(set(df_[column_name][i].strip().replace('КОМПОНЕНТЫ №', '').replace('(', '').replace(')', '').replace('.', '').replace('-', '').replace(',', '').split(' ')))
    try:
      all_[n_s].remove('')
      all_[n_s].remove('№')
      all_[n_s].remove('/')
      all_[n_s].remove(',')
      all_[n_s].remove('#')
    except ValueError as e:
      pass
      # print(e)
  for k, v in all_.items():
    for item in v:
      if item == ',':
        v.remove(item)
      elif item == '#':
        v.remove(item)
      try:
        float(item)
        v.remove(str(item))
        cnt += 1
      except Exception as e:
        pass
      if len(item) < 4:
        try:
          v.remove(str(item))
        except Exception as e:
          pass
      for char in item:
        if char.isdigit():
          try:
            v.remove(item)
          except Exception as e:
            pass
  # print(all_)
  return all_, [len(x) for x in all_.values()]

def get_roots(dict_, table=1, ch=[]):
  for k, v in dict_.items():
    n_l = []
    for item in v:
      item_n = stemmer.stem(item)

      # print(ch)
      if table == 2 and item_n in ch:
        n_l.append(item_n)
      elif table == 1:
        n_l.append(item_n)
      # else:
      #   print(item_n)
    dict_[k] = list(set(n_l))
  # print(dict_)

def get_values_pred(all_mas1, all_mas2, length):
  res = []
  count = []
  for key, value in all_mas2.items():
    di = {}
    for item in value:
      for key_, value_ in all_mas1.items():
        if item in value_:
          if key_ in di.keys():
            di[key_] += 1
          else:
            di[key_] = 1
    if di != {}:
      maxValue = max(di.values())
      if len([key for key in di.keys() if di[key]==maxValue]) == 1:
        res.append(max(di, key=di.get))
        count += list(di.values())
        continue
    res.append(NaN)
    count.append(NaN)
  while len(res) < length:
    res.append(NaN)
  # print(di)
  return res, count
```

```
def make_str(df, col):
  if col in df.columns:
    st = ', '.join([str(i) for i in df[col].unique() if pd.notna(i)])
    return st

def make_str_name(df, col):
  if col in df.columns:
    st = ' // '.join([str(i) for i in df[col].unique() if pd.notna(i)])
    return st

def get_tag(df):
  col = df['Таговый номер из E'].map(lambda x: re.sub(r'[^a-zA-Z0-9_\/]', r'', str(x)))
  if col.count() != 0:
    a = col[0].split('/')
    column = 'Артикул' # НоменклатураАртикул
    col = df[column].map(lambda x: x.split('/')[0] if pd.notna(x) else NaN)
    return col, a
  else:
    return None, None

def ch1(df):
  reg_list = ['commissioning spares', 'commis', 'start+up', 'Клей', 'Герметик', 'Краска', 'Масло ', 'Рем+комплект', 'Инструмент', 'Тележк', 'Лебедк', 'Лампа', 'Лампы', 'Лампочка', 'Лампочки', 'Телефон', 'Компьютер', 'Лаптоп',
      'Ноутбук', 'Программа', 'CD', 'Флэш', 'Планшет', 'Стилус', 'Рупор', 'Систем+громк+оповещ', 'Звук+оповещ', 'Поильник', 'Кабин', 'Кран', 'Запасн', 'Дополнитель']
  for key in reg_list:
    key_ = key.split('+')
    if len(key_) == 1:
      df.loc[df['Номенклатура'].str.contains(key, regex=False, case=False, na=False), 'Ch1_riskwords'] = f'в описании товара указано {key}'
    else:
      df.loc[(pd.concat([df['Номенклатура'].str.contains(word,regex=False, case=False, na=False) for word in key_],axis=1)).sum(1) > 1, 'Ch1_riskwords'] = f'в описании товара указано {key}'

def ch1_1(df):
  reg_list = ['Караск', 'Масл', 'Ламп', 'Программ', 'Диск', 'Душ ', 'start']
  for key in reg_list:
    df.loc[df['Номенклатура'].str.contains(key, regex=False, case=False, na=False), 'Ch1.1_riskwords'] = f'в описании товара указано {key}'
    
def ch4(df):
  if df['кол-во ед'].astype('int').sum() == df['Количество'].astype('int').sum():
    return '+'
  else:
    return 'проверить количество'

def ch5(df):
  if df['Номенклатура'].str.contains(r'[Шш]пильк').sum() > 4 or df['Номенклатура'].str.contains(r'[Гг]айк').sum() > 4:
    return 'проверить комплектность: гайки, шпильки'
  else:
    return NaN

def ch5_1(df):
  if df['номер строки_y'].count() > 5:
    return "проверить комплектность: количество позиций" 
  else:
    return NaN

def ch7(df):
  name_tov = df['наименование товара'].values[0].lower()
  if re.findall(r'стал', name_tov):
    if re.findall(r'нерж|коррозионностойк|дуплекс', name_tov):
      reg = r'[AaАа](182|403|312|182|351)'
    elif re.findall(r'углер', name_tov):
      reg = r'[AaАа](420|350|671|333|352|516)'
    else:
      return 'есть только корень "стал"'
    return df['Номенклатура'].map(lambda x: '+' if re.findall(reg, x) else ('указана неверная марка стали' if re.findall(r'[aAАа]\d{3}', x) else 'не указана марка стали'))
  return NaN

def ch7_1(df):
  name_tov = df['наименование товара'].values[0].lower()
  if re.findall(r'стал', name_tov):
    if re.findall(r'нерж|коррозионностойк|дуплекс', name_tov):
      reg = r'нерж|коррозионностойк|дуплекс'
    elif re.findall(r'углер', name_tov):
      reg = r'углер'
    else:
      return 'есть только корень "стал"'
    return df['Номенклатура'].map(lambda x: '+' if re.findall(r'стал', x) and re.findall(reg, x) else ('указана словами неверная марка стали' if re.findall(r'[aAАа]\d{3}', x) else 'не указана словами марка стали'))
  return NaN

def ch7_2(value):
  name_tov = value.lower()
  if re.findall(r'стал', name_tov):
    if re.findall(r'нерж|коррозионностойк|дуплекс', name_tov):
      reg = r'[AaАа](182|403|312|182|351)'
    elif re.findall(r'углер', name_tov):
      reg = r'[AaАа](420|350|671|333|352|516)'
    else:
      return NaN

    if re.findall(reg, name_tov):
      return '+'
    elif re.findall(r'[aAАа]\d{3}', name_tov):
      return 'в описании ПТИУ указана неверная марка стали'
    else:
      return 'в описании ПТИУ не указана марка стали'
  else:
    return NaN

def unnesting_all(df, explode, dt_name):  
  list_df = []
  for ind, row in df.iterrows():
    print(ind)
    blockt = row['Block_type']
    dt = row['index']
    row = row.drop(columns=['index'])
    if blockt not in [6, 7]:
      x = list(row['0_x'])
    if blockt != 5:
      y = list(row['0_y'])
    if blockt == 5 or (len(y) == 0 and len(x) != 0):
      df_ = pd.concat([pd.DataFrame(row[explode[0]]), pd.DataFrame([pd.NA], columns=['номер строки_y'])] , axis=1)
    elif blockt in [6, 7] or (len(y) != 0 and len(x) == 0):
      df_ = pd.concat([pd.DataFrame([pd.NA], columns=['номер строки_x']), pd.DataFrame(row[explode[1]])] , axis=1)
    elif len(x) == 0 and len(y) == 0:
      df_ = pd.DataFrame({'col1': [NaN], 'col2': [NaN]})   
    else:
      df_ = pd.concat([pd.DataFrame(row[x]) for x in explode] , axis=1)
    # вставка нужных столбцов
    df_.columns = ['first', 'second']
    df_.columns = df_.columns.astype('object')
    df_.insert(0, 'Block_type', row['Block_type'])
    df_.insert(0, 'Block ID', row['Block ID'])

    if row['0_x'] != NaN or len(row['0_x']) != 0:
      df_ = pd.merge(df_, massive_1, left_on='first', right_on='номер строки_x', how='left')
    if row['0_y'] != NaN or len(row['0_y']) != 0:
      df_ = pd.merge(df_, massive_4, left_on='second', right_on='номер строки_y', how='left')
    df_.drop(['first', 'second'], axis=1, inplace=True)
    # проверки
    if blockt in [1, 2, 3, 4]:
      if blockt in [1, 2, 3]:
        if df_['Номенклатура'].count() != 0:
          ch1(df_)
          ch1_1(df_)
      # СЕМАНТИКА
      all_mas1, count_1 = get_keywords(df_, 'наименование товара', 1)
      get_roots(all_mas1)
      x = list(all_mas1.values())
      ch = [j for i in x for j in i]
      all_mas2, count_4 = get_keywords(df_, 'Номенклатура', 2)
      get_roots(all_mas2, 2, ch)
      predict, count = get_values_pred(all_mas1, all_mas2, df_.shape[0])
      if blockt in [2,3]:
        df_['predict'] = predict
        df_['predict_2'] = df_['predict'].map(lambda x: df_[df_['номер строки_x'] == x]['Номер компонента из E'].item() if pd.notna(x) else NaN)

      else:
        col_tags, tag = get_tag(df_)
        #четкое совпадение и есть таговые номера
        if tag:
          df_['Ch2_tag'] = col_tags.map(lambda x: '+' if pd.notna(x) and re.sub(r'\W', r'', x.split('(')[0]) in tag else ('проверить соответствие таговых номеров' if pd.notna(x) else NaN))
          df_['Ch2.1_tag2'] = col_tags.map(lambda x: 'проверить второй таговый' if pd.notna(x) and re.findall(r'\(5', x) else '+')
          df_['Ch5_per1'] = ch5(df_)
          df_['Ch5.1_per2'] = ch5_1(df_)
        #четкое совпадение, но нет таговых номеров
        else:
          df_['Ch4_amount'] = ch4(df_)
        #четкое совпадение, но мы не обращаем внимание на таговый номер
        df_['Ch7_steelmark'] = ch7(df_)
        df_['Ch7.1_steelmark'] = ch7_1(df_)

        df_['count_roots_1'] = count_1[0]
        df_['count_roots_4'] = count_4
        df_['common_roots'] = count
        df_['Ch8_words'] = round(df_['common_roots']/df_['count_roots_1']*100, 2)
      #все блоки кроме ненайденных
      df_['Ch7.2_steelmark2'] = df_['Номенклатура'].map(lambda x:ch7_2(x) if pd.notna(x) else NaN) 
   
    #Вставка дубляжей
    if blockt == 3:
      df_.insert(df_.columns.get_loc("наименование товара") + 1, 'наименование товара_дубль', '//'.join(dt_name[dt][4]))
      df_.insert(3, 'К КР_дубль', ','.join(dt_name[dt][3]))
      lambda: df_.insert(3, 'Таговые номера', ','.join(dt_name[dt][2])) if pd.notna(dt_name[dt][2]) else NaN
      df_.insert(3, 'номера_инвойсов', ','.join(dt_name[dt][1]))   
      df_.insert(3, 'номера_компонентов', ','.join(dt_name[dt][0]))   
    else:
      df_.insert(df_.columns.get_loc("наименование товара") + 1, 'наименование товара_дубль', make_str_name(df_, 'наименование товара'))
      df_.insert(3, 'К КР_дубль', make_str(df_, '№ КР'))
      df_.insert(3, 'Таговые номера', make_str(df_, 'Таговый номер из E'))
      df_.insert(3, 'номера_инвойсов', make_str(df_, 'Номера инвойсов без даты из К'))   
      df_.insert(3, 'номера_компонентов', make_str(df_, 'Номер компонента из E'))   
    if blockt == 4:
      df_.insert(3, 'номера ТД', list(df_['№ ТД'].unique())[0]) 
    else:
      df_.insert(3, 'номера ТД', dt)  
    # Добавление в список блоков
    df_ = df_.loc[:, ~df_.columns.duplicated()]
    list_df.append(df_)
# Соединение блоков в единую таблицу
  df_1 = pd.concat(list_df)
  print('ok')
  return df_1.reset_index(drop=True)
```
